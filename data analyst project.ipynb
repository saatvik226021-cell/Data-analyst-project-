```python
import pandas as pd

# Step 1: Load your Excel file (change path if needed)
file_path = "C:\\Users\\LOQ\\Documents\\DDW-0000C-10.xlsx"
raw = pd.read_excel(file_path, header=None)

# Step 2: Inspect first 12 rows
raw_head = raw.head(12)
display(raw_head)
```

```python
# Cell 4: Auto-detect header row by finding row that contains "State" or "Table" or "Age"
header_row = None
for i in range(0, 12):
    row_text = " ".join([str(x).lower() if pd.notna(x) else "" for x in raw.iloc[i].tolist()])
    if ("state" in row_text and "dist" in row_text) or "age" in row_text or "table" in row_text:
        header_row = i
        break
header_row = 3 if header_row is None else header_row
print("Using header row:", header_row)
```

```python
#  Read file again using detected header row (keeps remaining rows as data)
df = pd.read_excel(file_path, header=header_row)
df.shape
```

```python
#  Show the columns detected and first rows
print("Columns:")
print(df.columns.tolist())
display(df.head())
```

```python
#  Clean column names: strip, replace newlines, lowercase
df.columns = [str(c).strip().replace("\n"," ").replace("/", " ").strip() for c in df.columns]
df.columns = [c if c != 'nan' else '' for c in df.columns]
df.head(2)
```

```python
# Cell 8: Quick column-name mapping (attempt) - show top candidate names
for i,c in enumerate(df.columns):
    print(i, ":", c)
```

```python
# Cell 9: Try to pick key columns (state, district, area/type, age groups and counts)
# We'll search for common words in column names and build safe column references
col_state = next((c for c in df.columns if 'state' in c.lower()), None)
col_dist = next((c for c in df.columns if 'dist' in c.lower()), None)
col_area_type = next((c for c in df.columns if 'total' in c.lower() or 'rural' in c.lower() or 'urban' in c.lower()), None)
print("Detected -> State:", col_state, "| Dist:", col_dist, "| Area/Type example:", col_area_type)
```

```python
# Cell 10: Identify numeric columns (likely population/institution counts)
# Heuristic: columns whose column name contains 'persons' 'males' 'females' 'total' or are numeric-like
num_candidates = [c for c in df.columns if any(k in c.lower() for k in ['persons','males','females','total','attending','count','age','population'])]
print("Numeric-like candidate columns (sample):", num_candidates[:12])
```

```python
# Cell 11: Convert candidate numeric columns to numeric (strip commas & coerce)
for c in num_candidates:
    df[c] = pd.to_numeric(df[c].astype(str).str.replace(",","").str.strip(), errors='coerce')
# show numeric columns count
display(df[num_candidates].head())
```

```python
# Cell 12: Drop columns that are fully NaN (clean)
df.dropna(axis=1, how='all', inplace=True)
print("Remaining columns:", len(df.columns))
```

```python
# Cell 13: Basic info & summary
print(df.info())
display(df.describe().T)
```

```python
# Cell 14: Clean/rename main region columns to safe names (if found)
if col_state:
    df = df.rename(columns={col_state: 'State'})
if col_dist:
    df = df.rename(columns={col_dist: 'District_Code'})
# If area/type column exists rename it
if col_area_type:
    df = df.rename(columns={col_area_type: 'Category_Type'})
print(df.columns.tolist())
```

```python
# Cell 15: Show a small cleaned sample (first 10 meaningful rows)
display(df[[c for c in df.columns][:10]].head(10))
```
